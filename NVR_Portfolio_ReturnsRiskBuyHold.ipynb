```python
# Import libraries
import math
from pandas_datareader import data as pdr
import yfinance as yfin
yfin.pdr_override()

import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
# Getting all the stock data (NVR, etc.)
stockSymbols = ['NVR','SPY','ITB','LEN','TOL','DHI','PHM']
#stockSymbols = ['NVR','SPY','ITB','JPM','GOOG','ATVI','NFLX']
```

```python
# Getting the stock start date
Stock_dt = '2021-01-01'

# Getting end date
end_date = '2021-12-30'
```

```python
# Getting today's date
#today = datetime.today().strftime('%Y-%m-%d')
#today
```

```python
# Number of assets
num_assets = len(stockSymbols)
```

```python
# creating a fn 
def getMyAssets(stock=stockSymbols, start=Stock_dt,end=end_date,col='Adj Close'):
    data = pdr.get_data_yahoo(stock, start = start, end=end)[col]
    return data
```

```python
# Getting the adj. close prices
my_stocks = getMyAssets(stockSymbols)
```

```python
my_stocks
```

```python
# calculating the simple daily returns
daily_simp_ret=my_stocks.pct_change(1)
daily_simp_ret
```

```python
daily_simp_ret
```

```python
# calculating the annual returns
annual_returns = daily_simp_ret.mean()*252
annual_returns
```

```python
# calculating annual risks
annual_risks = daily_simp_ret.std()*math.sqrt(252)
annual_risks
```

```python
# sorting the exp. annual risk data in descending order
sorted_ark_df = annual_risks.sort_values(ascending=False)
sorted_ark_df
```

```python
# sorting the exp. annual return data in descending order
sorted_ar_df = annual_returns.sort_values(ascending=False)
sorted_ar_df
```

```python
# plotting the sorted annualized return as a bar chart
plt.figure(figsize=(16,9))
plt.bar(sorted_ar_df.index,sorted_ar_df)
plt.ylabel('Annual Returns')
plt.ylabel('Assets')
plt.xticks(rotation=90)
plt.title('Annual returns of assets')
plt.show()
```

```python
sorted_ark_df
```

```python
# Creating a dataframe to store the annual returns, risks of the assets
df=pd.DataFrame()
df['Exp. Annual Returns']=annual_returns
df['Exp. Annual Risk']=annual_risks
df['Ticker']=df.index
df['ratio']=df['Exp. Annual Returns']/df['Exp. Annual Risk']

# sorting
df.sort_values(by='ratio',axis=0,ascending=False,inplace=False)
```

```python
# visualizing the exp. annual returns and risks
fig, ax = plt.subplots(figsize=(16,9))
plt.title('Expected Annual Returns vs Expected Annual Risks')
ax.scatter(df['Exp. Annual Risk'],df['Exp. Annual Returns'],c='DarkBlue')
ax.set_xlabel('Expected Annual Risks')
ax.set_ylabel('Expected Annual Returns')

# annotating the points on the plot
for idx, row in df.iterrows():
    ax.annotate(row['Ticker'],(row['Exp. Annual Risk'],row['Exp. Annual Returns']),c='green')
    
plt.show()
```

```python
# comparing assets having higher return and lower risk than NVR
row='NVR'
df.loc[(df['Exp. Annual Returns']>df['Exp. Annual Returns'][row])&(df['Exp. Annual Risk']<df['Exp. Annual Risk'][row])]
```

```python
# removing assets with relative high risk and relative low returns
remove_asset_list=[]
for ticker in df['Ticker'].values:
    no_better_asset=df.loc[(df['Exp. Annual Returns']>df['Exp. Annual Returns'][ticker])&(df['Exp. Annual Risk']<df['Exp. Annual Risk'][ticker])].empty
    if no_better_asset==False:
        remove_asset_list.append(ticker)

#remove_asset_list.remove('NVR')
# list containing stocks that are underperforming
remove_asset_list
```

```python
len(remove_asset_list)
```

```python
# new df with only high performing assets
df_new=df.drop(remove_asset_list)
df_new
```

```python
# redoing the portfolio calculation
assets=df_new.index
num_assets=len(assets)
```

```python
# calculating the new simple daily returns of the new df
daily_simp_ret = my_stocks[assets].pct_change(1)

# calculating the new annual return
annual_ret = daily_simp_ret.mean()*252
annual_ret
```

```python
# creating the weights for our portfolio
n = 1.0/float(num_assets)
weights=[n]*num_assets
weights=np.array(weights)
weights
```

```python
# calculating the covariance
cov_annual_matrix=daily_simp_ret.cov()*252
cov_annual_matrix
```

```python
# calculating the annual variance
port_variance=np.dot(weights.T,np.dot(cov_annual_matrix,weights))
port_variance
```

```python
# calculating the porfolio volatility /risk
port_volatility = np.sqrt(port_variance)
port_volatility
```

```python
# calculating the expected returns
port_returns_expected = np.sum(weights*annual_ret)

# expected returns %
print(str(round(port_returns_expected*100,2))+'%')
```
